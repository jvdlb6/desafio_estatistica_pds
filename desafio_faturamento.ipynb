```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.__version__
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
# Média das vendas

print(f"{df_faturamento['valor'].mean():.2f}")
```

```python
# Gráfico de Barras Verticais
import matplotlib.ticker as mticker

df_faturamento['mes_ref'] = pd.to_datetime(df_faturamento['data_ref']).dt.strftime('%Y-%m')
df_sorted = df_faturamento.sort_values('valor', ascending=False)
ax = df_sorted.plot(kind='bar', x='mes_ref', y='valor', figsize=(12, 6), legend=True)
ax.ticklabel_format(style='plain', axis='y')

ax.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f"{x:,.0f}".replace(",", "."))
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_faturamento = df_faturamento.sort_values(by='data_ref')
ax = df_faturamento.plot.line(x='data_ref', y='valor')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
